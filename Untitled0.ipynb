## Splitting train and test data

```python
X_train = df_conc.iloc[0:1455 , :].values
X_test = df_conc.iloc[1455: , :].values
y_train = df['SalePrice'].values
```

## Standardizing

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, -2:] = sc.fit_transform(X_train[:, -2:])
X_test[:, -2:] = sc.transform(X_test[:, -2:])
```

```python
len(X_train)
```

```python
len(X_test)
```

# Training the Models

### Training CatBoost on the Training set

```python
from catboost import CatBoostRegressor
classifier = CatBoostRegressor()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
```

```python
df_result = pd.DataFrame(y_pred, columns=['SalePrice'])
df_submission = dt['Id']
df_submission = pd.concat([df_submission, df_result], axis=1)
print(df_submission)
```

```python
df_submission.to_csv('results.csv', index=False)
```
